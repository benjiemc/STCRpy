```python
import stcrpy
```

```python
example_pdb_file = './8gvb.cif'
```

```python
# load a TCR structure to a TCR object
from stcrpy import load_TCRs

tcr_8gvb = load_TCRs(example_pdb_file)[0]
```

```python
tcr_8gvb
```

```python
tcr_8gvb.save()
tcr_8gvb.save(tcr_only=True)
```

```python
tcr_8gvb.calculate_docking_geometry(mode="cys")
```

```python
tcr_8gvb.score_docking_geometry()
```

```python
peptide_interactions = tcr_8gvb.profile_peptide_interactions()
peptide_interactions
```

```python
tcr_beta_peptide_interactions = peptide_interactions[peptide_interactions.domain == 'VB']
tcr_beta_peptide_interactions
```

```python
tcr_8gvb.get_interaction_heatmap(plotting_kwargs={"mutation_index": ['4-LEU', '7-GLY'], "save_as":'8gvb_example_interactions_heatmap.png'})
```
